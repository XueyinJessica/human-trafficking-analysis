# 读取数据

- 根据关键词搜索‘离婚’，‘拐卖’，‘民事案件’并选择基层法院，总共获得结果601条，显示600条，去重后共获得案例567条。

```python
from os import listdir
from os.path import isfile, join

mypath = 'data/'
pathes = [f for f in listdir(mypath) if not isfile(join(mypath, f))]
filepath = []
for path in pathes:
    filepath += [mypath+path+'/'+f for f in listdir(mypath+path) if isfile(join(mypath+path, f))]
    #filepath += [f for f in listdir(mypath+path) if isfile(join(mypath+path, f))]
```

# 文本解码

```python
# 文本解码有问题，不影响分析
text = []
for file in filepath:
    with open(file,encoding='GB18030', errors='ignore') as f:
        lines = f.read()
    text.append(lines)
```

```python
text[0]
```

```python
s_text = set(text)
text = list(s_text)
```

# 在文本中提取获取案例判罚的法院

```python
import re
res = []
for s in s_text:
    res.append(re.search(r"([\u4e00-\u9fff])+人民法院", s).group(0))
```

# 案例数量排序

```python
from collections import Counter
h = Counter(res)
rank = [e for e in h.items()]
rank.sort(key = lambda x:-x[1])
```

```python
rank
```

```python
# 有19个县市在数据库中搜索’离婚‘，’拐卖‘ 有4-5条记录。
for location, count in rank[0:20]:
    print('{}包含‘离婚’与‘拐卖’的判决文书条数为{}'.format(location,count))
```

```python
#目前网上发酵的丰县在这个数据库中可查到的结果为2条
for location, count in [e for e in rank if '江苏省丰县人民法院' in e[0]]:
    print('{}包含‘离婚’与‘拐卖’的判决文书条数为{}'.format(location,count))
```
